###Importamos mais uma biblioteca pra formatar o gráfico.
###Pra editar o formato do eixo "y", usou-se o termo "yaxis", onde "axis" (usa-se a abreviação "ax") se refere aos eixos "y" e "x". Quando o objeto era o eixo "y" usou-se a expressão "yaxis" (y + axis).
###Pra formatar os números no eixo "y", depois de referenciado o formato (StrMethodFormatter) usou-se a expressção {x:,.2f}
####"x" = valor formatado, sem relação com os eixos "x" e "y"
####"," = separador de milhares no padrão americano
####"." = ponto flutuante em duas casas decimais

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker 

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(12,5))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor por Unidade da Federação")
```

###Pra encontrar os dados de duas colunas, colocar duplo colchete, identificando que são duas colunas e não apenas uma com o nome das duas colunas.
###Um meio de identificar colunas que estão com os dados todos preenchidos: calcular a média dessas colunas, então só aparecerão como colunas válidas aquelas que estiverem totalmente preenchidas: dados.mean()  -  As colunas com lacunas em branco serão eliminadas da visualização.
###Reincluir a coluna com o nome dos estados que foi colocada fora dessa seleção das médias: colunas_usaveis.insert(0, "Unidade da Federação")
###Pra que isso funcione é necessário transformar a "coluna_usável" relacionada ao index em uma tabela mutável: colunas_usaveis = dados.mean().index.tolist()

```python
dados.head()
```

```python
dados.mean()
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis
```

###Transformamos as colunas que tinham média em uma lista, chamada "culunas usáveis". Depois indicamos que apenas essa "lista de colunas" deveria ser exibida (adicionada a coluna "Unidades da Federação", que foi transformada em índice (index)). 
####Por exemplo, quando pedimos pra que apenas os meses de agosto/2008 e Set/2008 fossem mostrados, usamos o código: dados[["2008/Ago", "2008/Set"]]. Agora dentro desse mesmo código vai estar inserida essa lista de colunas geradas, a "colunas usáveis" com o código:
####colunas_usaveis = dados.mean().index.tolist()
####colunas_usaveis.insert(0, "Unidade da Federação")
####colunas_usaveis

```python
usaveis = dados[colunas_usaveis]
usaveis.head()
```

###"colunas_usaveis" é uma lista, indicando qual parte (slice) dos dados devem ser mostrados. Agora nossa base de dados será "usaveis":
###usaveis = dados[colunas_usaveis]

```python
usaveis = usaveis.set_index("Unidade da Federação")
usaveis
```
